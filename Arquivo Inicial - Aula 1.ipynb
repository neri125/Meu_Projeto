### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python

import pandas

# Passo 4: Calcular os indicadores (faturamento e quantidade de produtos vendidos)
# Abrir a base de dados


caminho = r"C:\Users\teusn\Downloads\Vendas - Dez.xlsx"  #utilizo o r antes da string para informar para o python que \ não é para ele lê como comando, Ex \n , mas entender como \ mesmo

#A variável tabela recebe o valor da base de dados, vai ler a tabela em excel
tabela = pandas.read_excel(caminho)
# Ver as informações da base de dados
print(tabela)
display(tabela)



# Somar o faturamento de todos os produtos
# Somar a coluna de Valor Final

faturamento = tabela["Valor Final"].sum()


# Somar a quantidade de produtos
# Somar a coluna de Quantidade

qtde_produto = tabela["Quantidade"].sum()

print(faturamento)
print(qtde_produto)
```
