### 12. Faça um programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
ganho_hora = float(input("Digite o valor que você ganha por hora: "))
horas_mes = float(input("Digite o total de horas trabalhadas no mês: "))
```

### Calule o salário bruto (horas * salario por hora)

```python
salario = ganho_hora * horas_mes
print("O salário bruto é de: {}".format(salario))
```

### Calcule o desconto do IR(11% do salário bruto)


```python
ir = 0.11 * salario
print("Desconto do IR é de: {}".format(ir))
```

### Calcule o desconto do INSS (8% do salário bruto)

```python
inss = 0.08 * salario
print("Desconto do INSS é de: {}".format(inss))
```

### Calcule o desconto do sindicato (5% do salário bruto)

```python
sindicato = 0.05 * salario
print("Desconto com o sindicato é de: {}".format(sindicato))
```

### Calcule o salário líquido (salário bruto - descontos)

```python
salario_liquido = salario - ir - inss - sindicato
print("O salário liquido é de: {}".format(salario_liquido))
```

### 13. Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. (para simplificação nesse momento, não se preocupe em arredondar a quantidade de latas a serem compradas - vamos trabalhar isso em breve)

```python
metros_area_pintada = float(input("Digite a área em metros quadrado a ser pintada: "))

qtd_litros = metros_area_pintada / 3
qtd_latas = qtd_litros / 18
preco  = qtd_latas * 80.00

print("A quantidade de latas necessária é {} e o preço será de: {}".format(qtd_latas, preco))
```
