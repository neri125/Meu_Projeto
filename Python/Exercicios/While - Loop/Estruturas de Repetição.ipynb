#### 1. Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.

```python
nota = input("Digite sua nota")

#Enquanto for verdadeiro, faça: (loop infinito)
while True:
    if not nota:
        print("Valor inválido")
        break
    
    nota = input("Digite sua nota")
```

#### 2. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

```python


nome = input("Digite seu nome: ")
senha = input("Digite sua senha")

# Enquanto for verdadeiro faça o código abaixo dentro de um loop
while True:
    if senha == nome:
        print("Erro")
        nome = input("Digite seu nome: ")
        senha = input("Digite sua senha")
    else:
        break
```

#### 7. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o faturamento total (soma) e o faturamento médio por mês (média).

```python
faturamento = float(input("Digite o valor de faturamento: "))
valor = []

for i in range(4):
    valor.append(faturamento)
    faturamento = float(input("Digite o valor de faturamento"))

soma = sum(valor)
media = soma / (len(valor))

print("A soma do faturamento é de {}".format(soma))
print("A média do faturamento é de {}".format(media))
```

#### 8. Faça um programa que consiga categorizar a idade das equipes de uma empresa. Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da equipe varia entre 0 e 25 (jovem) ,26 e 60 (sênior) e maior que 60 (idosa); e então, dizer se a equipe é jovem, sênior ou idosa, conforme a média calculada.

```python
idade = input("Digite a sua idade")
idade_equipe = []
while True:
    if not idade:
        break
    else:
        idade = int(idade)
        idade_equipe.append(idade)
        idade = input("Digite a sua idade")

media = (sum(idade_equipe)) / (len(idade_equipe))
media = float(media)

if media >= 0 and media <= 25:
    print("A equipe é jovem")

elif media >= 26  and media <= 60:
    print("A equipe  é sênior")

elif media > 60:
    print("A equipe é idosa")
```

#### 9. Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
total_eleitores = input("Total de eleitores? ")
voto = 0
candidato = []

while True:
    if total_eleitores:
        total_eleitores = int(total_eleitores)
        voto = voto + 1

        if voto <= total_eleitores:
            qual_candidato = int(input("Temos 3 candidatos, digite 1, ou 2 , ou 3"))
            candidato.append(qual_candidato)
        
        else:
            break

    else:
        break


candidato_1 = candidato.count(1)
candidato_2 = candidato.count(2)
candidato_3 = candidato.count(3)

print("O total de votos no candidato 1: {}".format(candidato_1))
print("O total de votos no candidato 2: {}".format(candidato_2))
print("O total de votos no candidato 3: {}".format(candidato_3))
```

#### 10. Faça um programa que calcule o valor total investido por um colecionador em sua coleção de CDs e o valor médio gasto em cada um deles. O usuário deverá informar a quantidade de CDs e o valor para em cada um.

```python
qtd_cds = int(input("Digite total de cds: "))
qtd_atual = 0
vendas = []

while True:
    qtd_atual = qtd_atual + 1
    if qtd_atual <= qtd_cds:
        valor = float(input("Digite o valor do CD:"))
        vendas.append(valor)

    else:
        break

valor_total = sum(vendas)
valor_medio = valor_total / (len(vendas))

print("O valor total do CD é de {}".format(valor_total))
print("O valor médio é de {}".format(valor_medio))
```
