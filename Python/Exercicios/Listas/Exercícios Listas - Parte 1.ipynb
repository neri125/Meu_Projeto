## 1. Faturamento do Melhor e do Pior Mês do Ano

Qual foi o valor de vendas do melhor mês do Ano?
E valor do pior mês do ano?

```python
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]

vendas_valor_final = vendas_1sem + vendas_2sem
print(vendas_valor_final)

#Sendo o valor 
maior = max(vendas_valor_final)
menor = min(vendas_valor_final)

#Sabendo a posição dos meses

i_maior = vendas_valor_final.index(maior)
i_menor = vendas_valor_final.index(menor)

print("O melhor mês foi {} com o valor de {:,} unidades vendidas \nO pior mês foi {} com o valor de {:,} unidades vendidas".format(meses[i_maior], maior, meses[i_menor], menor))



```

## 2. Continuação

Agora relacione as duas listas para printar 'O melhor mês do ano foi {} com {} vendas' e o mesmo para o pior mês do ano.

Calcule também o faturamento total do Ano e quanto que o melhor mês representou do faturamento total.

Obs: Para o faturamento total, pode usar a função sum(lista) que soma todos os itens de uma lista

```python
faturamento_total = sum(vendas_valor_final)
representacao = maior / faturamento_total

print("O melhor mês do ano foi {} com {} vendas".format(meses[i_maior], maior))
print("O total de faturamento foi R${:,}".format(faturamento_total))
print("O melhor mês do ano {} teve uma representação {:.2%} do faturamento total".format(meses[i_maior], representacao))
print("O pior mês do ano foi {} com {} vendas".format(meses[i_menor], menor))

```
